```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import RandomizedSearchCV, train_test_split
from sklearn.metrics import mean_absolute_error
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
```

```python
x = data.drop('CO2', axis=1)
y = data['CO2']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
scaler = StandardScaler()
scaled =scaler.fit_transform(x_train)
clf = LinearRegression()
clf.fit(x_train, y_train)
clf.score(x_test,y_test)
```

```python
y_preds = clf.predict(x_test)
mean_absolute_error(y_test, y_preds)
```

```python

x = data.drop('CO2', axis=1)
y = data['CO2']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
clf = RandomForestRegressor(n_estimators=20,max_depth=5)


clf.fit(x_train, y_train)
clf.score(x_test,y_test)
```

```python
y_preds = clf.predict(x_test)
mean_absolute_error(y_test, y_preds)
```

```python
x = data.drop('Temp', axis=1)
y = data['Temp']
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2)
Temp_clf =RandomForestRegressor(n_estimators=100,max_depth=50)
print(Temp_clf.fit(x_train, y_train))
print(Temp_clf.score(x_test,y_test))
```

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor(max_depth=50)
model.fit(x_train,y_train)

print(model.score(x_test,y_test))
```
